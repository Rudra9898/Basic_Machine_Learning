1. Use iris flower dataset from sklearn library and try to form clusters of flowers using petal width and length features. Drop other two features for simplicity.
2. Figure out if any preprocessing such as scaling would help here
3. Draw elbow plot and from that figure out optimal value of k

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
scaler=MinMaxScaler()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
scaler.fit(df)
df=scaler.transform(df)
```

```python
df
```

```python
df1=iris.target                      #pd.Series(iris.target)
df1
```

```python
scaler1=MinMaxScaler()
scaler1.fit(df1.reshape(-1, 1))
df1=scaler1.transform(df1.reshape(-1, 1))
df1
```

```python
SSE=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(df,df1)
    SSE.append(km.inertia_)
```

```python
plt.plot(SSE)
```

```python
# here found cluster=2
```

```python
km1=KMeans(n_clusters=2)
km1.fit(df,df1)
y_pred=km1.predict(df)
y_pred
```

```python
km2=KMeans(n_clusters=2)
y_pred1=km2.fit_predict(df,df1)
```

```python
y_pred1
```

```python
plt.scatter(df[:,0],y_pred1,color='r',label=iris.feature_names[0])
plt.scatter(df[:,1],y_pred1,color='g',label=iris.feature_names[0])
plt.scatter(df[:,2],y_pred1,color='b',label=iris.feature_names[0])
plt.scatter(df[:,3],y_pred1,color='m',label=iris.feature_names[0])
plt.legend()
plt.show()
```

```python
df=pd.DataFrame(df,columns=iris.feature_names)
df['predicted_target']=y_pred
```

```python
df.head()
```

```python
df1 = df[df.predicted_target == 0] 
df2 = df[df.predicted_target == 1]  

plt.figure(figsize=(10, 6)) 

plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], label='Cluster 0', color='red')
plt.scatter(df1['petal length (cm)'], df1['petal width (cm)'], label='Cluster 0', color='red', alpha=0.3)  # Add transparency (optional)

plt.scatter(df2['sepal length (cm)'], df2['sepal width (cm)'], label='Cluster 1', color='green')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], label='Cluster 1', color='green', alpha=0.3)

plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Iris Dataset - Scatter Plot by Predicted Cluster')
plt.legend()
plt.grid(True)  # Add grid lines (optional)
plt.show()
```
